```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
data=pd.read_csv("diabetes.csv")
df=data.copy()
df.head()
```

```python
degiskenler = [
    'GebelikSayısı',
    'Glikoz',
    'KanBasıncı',
    'CiltKalınlığı',
    'İnsülin',
    'VücutKitleİndeksi',
    'DiyabetSoyAğacıFonksiyonu',
    'Yaş',
    'Sonuç'
]
df.columns=degiskenler
df.head()
```

```python
hasta=df[df["Sonuç"]==1]
sağlikli=df[df["Sonuç"]==0]
```

```python
plt.scatter(hasta.Yaş,hasta.Glikoz,label="hasta",alpha=0.4,color="red")
plt.scatter(sağlikli.Yaş,sağlikli.Glikoz,label="sağlikli",alpha=0.4,color="green")
plt.xlabel("yaş")
plt.ylabel("Gilikoz oranı")
plt.legend()
plt.show();
```

```python
sns.barplot(x="GebelikSayısı", y="Sonuç", data=df)
plt.show()
```

```python
sns.scatterplot(x="GebelikSayısı", y="Glikoz", data=df)
plt.show()
```

```python
y=df["Sonuç"].values

```

```python
x=df.drop(["Sonuç"],axis=1)
x
```

```python
z = (x - x.min()) / (x.max() - x.min())
z
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=1)
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(xtrain,ytrain)
prediction=knn.predict(xtest)
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)

print("k=3 için test verilerimizin doğrulama testi sonucu : ",knn.score(xtest,ytest))
```
